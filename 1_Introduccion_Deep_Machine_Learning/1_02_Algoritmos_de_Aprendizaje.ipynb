# <a name="M1">1. Algoritmos de Aprendizaje</a>


* Los ***Algoritmos de Aprendizaje, son unos Algoritmos que reciben como entrada una serie de datos*** (etiquetados para el aprendizaje supervisado y no etiquetados para el aprendizaje no supervisado) y tienen la ***misión de aprenden de manera autónoma el conocimiento que hay implicito en los datos, representando ese conocimiento en un modelo que es la salida del Algoritmo de Aprendizaje***, con el fin de usar ese modelo para realizar predicciones en el futuro con nuevos datos.


<img src="../imgs/1_18_ml.png" style="width: 600px;"/>


* Por ejemplo para el Aprendizaje Supervisado, podemos poner como ejemplo que un Algoritmos de Aprendizaje en particular va a tener que encontrar la mejor función $f(X)$ que mejor relacione las variables de entrada $X$ con una variable de salida $y$, siendo $f(x)$ el resultado del Algoritmo de Aprendizaje; es decir, el modelo:


$$y= f(X)$$


* En función de los datos que tengamos y del problema que queramos resolver, disponemos de diferentes Algoritmos de Aprendizaje y de diferentes técnicas de evaluación para crear los modelos. En la siguiente imagen podemos ver los Algoritmos de Aprendizaje más conocidos para cada uno de los tipos de aprendizaje y técnicas:


<img src="../imgs/1_16_ml.png" style="width: 1200px;"/>


* Cada uno de estos Algoritmos de Aprendizaje tendrá sus particularidades a la hora de aprender y de generar el modelo. En función de estas particularidades podremos clasificar a los Algoritmos de Aprendizaje de una manera o de otra, como veremos en los siguientes puntos.

# <a name="M2">2. Algoritmos de Aprendizaje Paramétricos y no Paramétricos</a>


* Una de las maneras de clasificar los Algoritmos de Aprendizaje es en Algoritmos de Aprendizaje Paramétricos y No Paramétricos.


### - Algoritmos de Aprendizaje Paramétricos


* Los ***Algoritmos de Aprendizaje Paramétricos*** son aquellos que ***su modelo lo construyen basandose en una función o forma conocida***; por ejemplo, la Regresión Lineal construye un modelo de la forma $y = \beta_0 + \beta_1 \cdot X$ y el objetivo de este Algoritmo de Aprendizaje es encontrar los mejores parámetros $\beta_0$ y $\beta_1$ que mejor se ajusten a los datos. 


* Algunos ejemplos de Algoritmos de Aprendizaje Paramétricos son:

    - Regresión Lineal
    - Regresión Logística
    - Linear Discriminant Analysis
    
    
* Veamos a continuación las ventajas e inconvenientes de este tipo de Algoritmos de Aprendizaje:

    - ***Ventajas***:
        + ***Simplicidad***: Estos algoritmos son sencillos de entender e interpretar.
        + ***Velocidad***: Son algoritmos que aprenden bastante rápido.
        + ***Pocos Datos***: Este tipo de algoritmos necesita pocos datos para generar un buen modelo.
    
    - ***Inconvenientes***:
        + ***Limitaciones***: Estos modelos estan restringidos a la forma especificada.
        + ***Complejidad baja***: Este tipo de algoritmos estan bien para resolver problemas simples ya que generan modelos poco complejos.


### - Algoritmos de Aprendizaje No Paramétricos


* Los ***Algoritmos de Aprendizaje No Paramétricos*** son aquellos que generan ***un modelo que no tiene (a priori) una forma conocida***.


* Algunos ejemplos de Algoritmos de Aprendizaje No Paramétricos son:

    - Arboles de Decisión
    - Naive Bayes
    - Support Vector Machine - SVM
    - Redes Neuronales
    
    
* Veamos a continuación las ventajas e inconvenientes de este tipo de Algoritmos de Aprendizaje:

    - ***Ventajas***:
        + ***Flexibilidad***: Generan modelos que pueden tener muchas formas diferentes.
        + ***Eficiencia***: Al generar modelos más complejos, estos suelen dar mejores resultados de acierto para problemas complejos.
    
    - ***Inconvenientes***:
        + ***Muchos Datos***: Este tipo de algoritmos necesita muchos datos para generar un buen modelo.
        + ***Lentos***: Estos algoritmos aprenden muy lentamente y tardan bastante en generar un buen modelo.
        + ***Sobreajuste (Overfitting)***: Este tipo de algoritmos tiende a generar modelos que se ajustan bastante a los datos de entrenamiento.

# <a name="M3">3. Algoritmos de Aprendizaje de Caja Blanca y Caja Negra</a>


* Otra de las maneras que tenemos de clasificicar los diferentes tipos de Algoritmos de Aprendizaje es en función de la ***interpretabilidad del modelo generado por parte de un humano***.


* Uno de los retos que tenemos a la hora de realizar predicciones para la toma de decisiones tras explotar un modelo, es ***poder explicar o justificar la predicción realizada***.


* Por lo general los modelos "simples" suelen ser modelos interpretables pero de baja eficiencia, mientras que los modelos "complejos" no suelen ser interpretables pero ofrencen muy buenos resultados.


* Por tanto podemos dividir a los algoritmos de aprendizaje en:

    + ***Algoritmos de de Aprendizaje de Caja Blanca***: Algoritmos de Aprendizaje que generan modelos cuyas predicciones se pueden justificar; como por ejemplo, los arboles de decisión o los algoritmos que generan modelos probabilísticos.
    
    + ***Algoritmos de de Aprendizaje de Caja Negra***: Algoritmos de Aprendizaje que generan modelos cuyas predicciones no se pueden justificar de una manera más o menos razonable; como por ejemplo, las redes neuronales, SVM, etc.
