## <a name="M1">1. Clustering Jerárquico</a>


* Los ***método Jerárquicos para la resolución de problemas de Clustering, son métodos que tienen como objetivo agrupar clusters para formar uno nuevo o separar algún Cluster ya existente para dar origen a otros dos clusters***. Estas acciones se repiten hasta que todos los elementos queden en un único Cluster o hasta que cada elemento sea un cluster respectivamente.

* Este típo de métodos en realidad no muestran una Clusterización como el K-Means o el Gaussian Mixture Models, si no que lo que ***muestran son las relaciones de proximidad o similaridad entre elementos o subconjuntos de elementos***.


* ***Estudiando estas relaciones de similaridad entre elementos, podremos ser capaces de seleccionar el número de Clusters que queramos***.


* Los métodos jerárquicos los podemos clasificar en:
<span></span><br><br>
    + ***Aglomerativos***: Inicialmente se forman tantos Clusters como elementos haya (Clusters de 1 elemento). En cada iteracion se unen los dos clusters más próximos hasta que todos los elementos formen un Cluster.
<span></span><br><br>
    + ***Divisivos***: Inicialmente todos los elementos pertenecen a un único Cluster. En cada iteración se selecciona un Cluster y se divide en dos hasta que haya tantos Clusters como elementos.
    
    
* La ventaja de este tipo de métodos jerárquicos es que no es necesario definir a priori el número de Clusters.

## <a name="M2">2. Clustering Jerárquico Aglomerativo</a>


* Los ***métodos Jerárquicos Aglomerativos para la resolución de problemas de Clustering, son métodos que consisten en agrupar los elementos en subconjuntos cada vez mayores, hasta que todos los elementos conformen un único Cluster***.


* Esta agrupación de elementos o subconjuntos de elementos se realiza viendo la similaridad (o proximidad) entre elementos o subconjuntos de elementos utilizando alguna métrica de similaridad como pueda ser por ejemplo la distancia euclidea:
    
$$d(x,y) = \sqrt{\sum_{i=1}^{N}(x^{(i)} - y^{(i)})}$$


### Dendrogramas

* ***La manera de representar un Clusterin Jerárquico es con un Dendrograma que es un tipo de representación gráfica en forma de árbol que organiza y agrupa los elementos en subcategorías según su similitud. Los objetos similares se representan en el dendrograma por medio de un enlace cuya posición está determinada por el nivel de similitud entre los elementos o grupos de elementos***.

### Pros y Contras del Clustering Jerárquico Aglomerativo


* ***CONTRAS***: Como puede observarse la carga computacional de este método es muy alta ya que tiene que calcular las distancias de "todos con todos" los elementos y los subconjunto de elementos formados, por lo que no es una técnica muy apropiada para usarla con Datasets muy grandes.


* ***PROS***: La ventaja de los métodos Jerárquicos Aglomerativos frente a Algoritmos de Aprendizaje como el K-Means o el Gaussian Mixture Models es que no hay que definir previamente el número de clusters ya que este lo podemos obtener viendo las agrupaciones realizadas.

## <a name="M3">3. Ejemplo sencillo de Clustering Jerárquico Aglomerativo</a>


* Para resolver el problema de Clustering Jerárquico Aglomerativo vamos a utilizar la clase **"AgglomerativeClustering()"**, cuya documentación esta en el siguiente enlace:<br>
https://scikit-learn.org/stable/modules/generated/sklearn.cluster.AgglomerativeClustering.html


* Para construir y visualizar el Dendrograma, vamos a usar la clase **"dendrogram"** de la librería scipy, cuya documentación esta en el siguiente enlace:<br>
https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.cluster.hierarchy.dendrogram.html


* Este ejemplo los vamos a resolver con los puntos a Clusterizar:

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 

# Leemos los datos
X = np.loadtxt('../../data/DS_3Clusters_999Puntos.txt', delimiter='::')

# Pintamos los elementos
plt.scatter(X[:,0], X[:,1], color='c', marker='o')
plt.xlabel('X1')
plt.ylabel('X2')
plt.plot()
```

* Una vez vistos los elementos, vamos a construir el Dendrograma, indicando como metrica de similaridad la distancia euclidea:

```python
from scipy.cluster.hierarchy import dendrogram, linkage

# Calculamos las distancias entre puntos o grupos de puntos
Z = linkage(X, metric='euclidean', method='ward')

# Generamos el Dendrograma y lo pintamos
dendrogram(
        Z,
        truncate_mode='lastp',
        p=15,
        leaf_rotation=90.,
        leaf_font_size=12.,
        show_contracted=True
    )
plt.title('Dendrograma')
plt.xlabel('Elementos')
plt.ylabel('Distancia Euclidea')
plt.show()
```

* Una vez que tenemos el Dendrograma, podemos observar como realizando una poda (o corte) en el nivel 2, tendríamos 3 grupos de elementos bien diferenciados, asi que vamo a pasar a realizar un Clustering Jerárquico Aglomerativo para 3 Clusters.

```python
# Ajustar el clustetring jerárquico a nuestro conjunto de datos
from sklearn.cluster import AgglomerativeClustering
agc = AgglomerativeClustering(n_clusters = 3, affinity = "euclidean", linkage = "ward")
agc.fit(X)
```

* Pasamos a continuación a mostrar los resultados de la Clusterización:

```python
# Obtenemos las etiquetas de cada uno de los puntos
y_pred = agc.labels_

# Pintamos los elementos en su cluster correspondiente diferenciado por color:
COLORS = ['red', 'blue', 'green', 'gray', 'violet', 'brown', 'cyan', 'magenta']
fig, ax =  plt.subplots(figsize=(8, 6))
ax.set_title('Clusters')
ax.set_xlabel('X1')
ax.set_ylabel('X2')

for cluster in range(agc.n_clusters_):
    plt.scatter(X[y_pred == cluster, 0], X[y_pred == cluster, 1], color=COLORS[cluster], marker='.')
plt.plot()
```
